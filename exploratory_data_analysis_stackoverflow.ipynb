```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

#### Handling Missing Data


```python

# Check missing values
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())

# Fill categorical columns with mode
df['Employment'] = df['Employment'].fillna(df['Employment'].mode()[0])
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])

# Drop rows where JobSat is missing
df = df.dropna(subset=['JobSat'])

# Verify
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())
```

#### Analysis of Experience and Job Satisfaction


```python

# Convert YearsCodePro to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Create experience ranges
bins = [0, 5, 10, 20, float('inf')]
labels = ['0-5', '5-10', '10-20', '>20']

df['ExperienceRange'] = pd.cut(
    df['YearsCodePro'],
    bins=bins,
    labels=labels
)

# Median JobSat per experience range
median_jobsat = df.groupby('ExperienceRange')['JobSat'].median()

print(median_jobsat)

# Visualization
median_jobsat.plot(kind='bar')
plt.title('Median Job Satisfaction by Experience Range')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction')
plt.show()
```

#### Visualize Job Satisfaction


```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='JobSat')

plt.title('Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')

plt.show()
```

#### Analyzing Remote Work Preferences by Job Role


```python
# Remote work distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='RemoteWork')

plt.title('Remote Work Preferences')
plt.xticks(rotation=20)
plt.show()

# Employment vs RemoteWork
remote_emp = pd.crosstab(
    df['Employment'],
    df['RemoteWork']
)

print(remote_emp)

# Top employment categories only (for readability)
top_emp = df['Employment'].value_counts().head(10).index

plt.figure(figsize=(12,6))
sns.countplot(
    data=df[df['Employment'].isin(top_emp)],
    y='Employment',
    hue='RemoteWork'
)

plt.title('Remote Work Preferences by Employment Type')
plt.show()
```

#### Analyzing Programming Language Trends by Region


```python
lang_region = df[['Country', 'LanguageHaveWorkedWith']].dropna()

lang_region['LanguageHaveWorkedWith'] = (
    lang_region['LanguageHaveWorkedWith']
    .str.split(';')
)

lang_region = lang_region.explode('LanguageHaveWorkedWith')
```

```python
language_by_country = (
    lang_region
    .groupby(['Country', 'LanguageHaveWorkedWith'])
    .size()
    .unstack(fill_value=0)
)

print(language_by_country.head())
```

```python
plt.figure(figsize=(14, 7))
sns.heatmap(language_by_country, cmap='Blues')
plt.title('Top Programming Languages by Country')
plt.xlabel('Programming Language')
plt.ylabel('Country')
plt.show()
```

#### Correlation Between Experience and Satisfaction


```python
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['JobSatPoints_1'] = pd.to_numeric(df['JobSatPoints_1'], errors='coerce')

plt.figure(figsize=(8, 5))
plt.scatter(df['YearsCodePro'], df['JobSatPoints_1'], alpha=0.5)
plt.title('Correlation Between Experience and Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Points')
plt.show()

correlation = df[['YearsCodePro', 'JobSatPoints_1']].corr()
print(correlation)
```

#### Educational Background and Employment Type


```python
edu_emp = pd.crosstab(df['EdLevel'], df['Employment'])

print(edu_emp)

plt.figure(figsize=(12, 6))
sns.heatmap(edu_emp, cmap='Blues')
plt.title('Education Level by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Education Level')
plt.show()
```
